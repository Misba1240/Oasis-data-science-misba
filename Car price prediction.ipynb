```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
car = ['Car_Name','Year','Selling_Price','Present_Price','Driven_kms','Fuel_Type','Selling_type','Transmission','Owner']
for var in car:
    label_encoder = LabelEncoder()
    df[var+'_label_encoded'] = label_encoder.fit_transform(df[var])
df.drop(car,axis=1,inplace = True)
print("The encoded data\n:")
df
```

```python
#we must drop the price because we have to predict it
x = df.drop(columns =['Selling_Price_label_encoded','Car_Name_label_encoded'],axis=1)
y = df['Selling_Price_label_encoded']
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
```

```python
model1.fit(X_train,Y_train)
```

```python
#training set prediction
training_data_prediction = model1.predict(X_train)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
error_score = metrics.r2_score(Y_train,training_data_prediction)
print("R squared error :",error_score)
```
